None of the intermediate objects in a method chain produces an output, so if one of the methods produces an unexpected or unusable result, it will be hard to trace where in the chain a bug occured.

# Scaling a trailing stock order using cummax method #

GOing to use pd.data_reader to simplify extraction process and pre-processing, before using another sorting method to figure out when the best time to sell a stock is

````python
import pandas_datareader as pdr
````

````python
tsla = pdr.DataReader('tsla', data_source='yahoo', 
                      start = '2017-1-1')

tsla.head(8)
````

Want to keep this simple, so let's only work with the closing price when setting the stop order

````python
tsla_close = tsla['Close']
````

Use cummax() function to track highest closing price until end of dataset

````python
tsla_cummax = tsla_close.cummax()

tsla_cummax.head(8)
````

````python
tsla_trailing_stop = tsla_cummax * 0.9
````

````python
tsla_trailing_stop.head(8)
````

Could even pass this into a function that returns a trailing stop on demand

````python
def set_trailing_loss(symbol, purchase_date, perc):
    close = pdr.DataReader(symbol, 'yahoo', 
                           start = purchase_date)['Close']
    return close.cummax() * perc
````

````python
msft_trailing_stop = set_trailing_loss('msft', '2017-6-1', 0.85)
msft_trailing_stop.head(8)
````

## Constructing Multiple Boolean Conditions ##

In this example, we are going to filter for movies with IMDB scores > 8, content rating of PG-13, and a title_year before 2000 or after 2009

In Pandas, boolean operators are ```|``` for ```or```, ambersand for ```and```, ```-``` for ```not```.

Means that thse comparisons aren't as readable in Pandas, but Python comparasion operations evaluate a series as a whole, whereas Pandas operators eveluate each element in a series or dataframe. 

````python
criteria1 = movie.imdb_score > 8

criteria2 = movie.content_rating == 'PG-13'

criteria3 = ((movie.title_year < 2000) |
             (movie.title_year > 2009))

criteria_final = criteria1 & criteria2 & criteria3

criteria_final.head()
````

## Constructing DataFrame Filters

````python
#Construct first filter set (same as before)

crit_a1 = movie.imdb_score > 8
crit_a2 = movie.content_rating == 'PG-13'
crit_a3 = ((movie.title_year < 2000) | (movie.title_year > 2009))

final_crit_a = crit_a1 & crit_a2 & crit_a3

#Now filter for movies with scores below 5, rated R, and released during the naughties

crit_b1 = movie.imdb_score < 5
crit_b2 = movie.content_rating == 'R'
crit_b3 = ((movie.title_year >= 2000) & (movie.title_year <= 2010))

final_crit_b = crit_b1 & crit_b2 & crit_b3

#Now evaluate movies that fit either filter A or B as one boolean condition

final_crit_all = final_crit_a | final_crit_b
final_crit_all.head()
````

````python
movie[final_crit_all].head()
````

````python
cols = ['imdb_score', 'content_rating', 'title_year']

movie_filtered = movie.loc[final_crit_all, cols]

movie_filtered.head(10)
````

# Determining Normality of Stock Returns #

````python
amzn = pd.read_csv('data/amzn_stock.csv', index_col = 'Date', parse_dates = ['Date'])

amzn.head()
````

````python
amzn_daily_return = amzn.Close.pct_change()

amzn_daily_return.head()
````

````python
amzn_daily_return = amzn_daily_return.dropna()

amzn_daily_return.hist(bins=20)
````

The routine we just followed allowed us to see the daily return on Amazon stock via the ```pct_change()``` method, which was then filtered for NA's before being passed into a histogram. Now let's analyze the distribution for normality. 

````python
mean = amzn_daily_return.mean()

std = amzn_daily_return.std()

abs_z_score = amzn_daily_return.sub(mean).abs().div(std)
````

Now we find the percentage of returns within 1,2,and 3 standard deviations

Below is a really great example of a simple for loop!

Keep in mind in reocrds

````python
pcts = [abs_z_score.lt(i).mean() for i in range(1,4)]

print('{:.3f} fall within 1 standard deviation. '
      '{:.3f} within 2 and {:.3f} within 3'.format(*pcts))
````

## The above operation is crucial ## 

We can pass values of an expression into a list, and call them in a print statement with the .format() method.

```{:.3f}``` is telling python to access the list in .format(), which by being preceded by an asterikc, causes each respective value in the lsit to eb called at each in stance of the {:.3f} call.  {:.3f} itself tells python to import the values to 3 points of rpecision, as a floating point numeric value. 

## We can write this normalization checking procedure as a function ##

````python
def test_return_normality(stock_data):
    #Call Closing stock prices and perform relevant data filtering, tabulation, and graphing functions
    close = stock_data['Close']
    daily_return = close.pct_change().dropna()
    daily_return.hist(bins=20)
    mean = daily_return.mean()
    std = daily_return.std()
    # Calculate z score and percentage of closing vlaues that fall within a given std range
    abs_z_score = abs(daily_return - mean) / std
    pcts = [abs_z_score.lt(i).mean() for i in range(1,4)]
    #Return a pritn statement specifying normality distribution from abs_z_score list
    print('{:.3f} fall within 1 standard deviation. '
      '{:.3f} within 2 and {:.3f} within 3'.format(*pcts))
````

````python
slb = pd.read_csv('data/slb_stock.csv', index_col = 'Date', parse_dates = ['Date'])

test_return_normality(slb)
````

## Customizing an Aggregate Function 

Create your own Aggregate function for a more complex summary on a given column

Going to create aggregate function that finds how many standard deviation unitsa away from the mean a given state's undergraduate population is distributed

Now we define function, which is effectively a standardized population score, which we then take the max of for a given state

````python
def max_deviations(s):
    std_score = (s - s.mean()) /s.std()
    return std_score.abs().max()
````

## Customizing aggregating functions with *args and **kwargs functions

By default, Pandas only passes each of the aggregating columns to an aggregating function **one at a time AND as a Series**

Python has the ability to pass an arbitrary number of arguments to a given function, and sometimes we need more passed than just a Series itself.

Let's see what we mean by this by using the `inspect` module on the `agg` function

````python
college = pd.read_csv('data/college.csv')
````

````python
grouped = college.groupby(['STABBR', 'RELAFFIL'])
````

````python
import inspect
inspect.signature(grouped.agg)
````

Going to build a custom function that finds the percentage of schools, grouped by state and religious association, that have an undergraduate population between two values. 

We can actually define our function in such a way to allow for multiple args during the `.agg()` method

````python
def pct_between(s, low, high):
    return s.between(low, high).mean()
````

````python
college.groupby(['STABBR', 'RELAFFIL'])['UGDS']\
       .agg(pct_between, 1000, 10000).head(9)
````

Function we defined, `pct_between`, is very flexible. By calling two non-keyword *args* the the aggregate function, Pandas passses thnose scalars into the method. 

In order to use more than one custom function, or a custom function and a standard function, need to define a special function called a `closure`

````python
def make_agg_func(func, name, *args, **kwargs):
    def wrapper(x):
        return func(x, *args, **kwargs)
    wrapper.__name__ = name
    return wrapper
````

````python
my_agg1 = make_agg_func(pct_between, 'pct_1_3k', low = 1000, high = 3000)
my_agg2 = make_agg_func(pct_between, 'pct_10_30k', low = 10000, high = 30000)
````

Now we are ready to pass these functions in a grouping/aggregate method. 

Don't place closed functions in quotes

````python
college.groupby(['STABBR', 'RELAFFIL'])['UGDS']\
       .agg(['mean', my_agg1, my_agg2]).head()
````

## Calculating Weighted Mean SAT scores per state with apply method

We've now met the four `.groupby` methods that accept a function: `.agg`, `.filter`, `.transform`, and `.apply`

.agg must return a scalar value

.filter must return a boolean

.transform must return a series with same length as passed group

However, .apply can do any of these things. Furthermore, it gets passed once per group, as opposed to .agg and .transform which only get passed for non-grouped columns. In the end, .apply() allows us to be more flexible

To perform stated operation, need to first filter data so that desired columns don't have NaN's:

````python
college = pd.read_csv('data/college.csv')
````

````python
subset = ['UGDS', 'SATMTMID', 'SATVRMID']
````

````python
college2 = college.dropna(subset=subset)
````

````python
college.shape
````

````python
college2.shape
````

Next we define our custom function to calculate weighted math score, and then pass it to the .apply() method:

````python
def weighted_math_average(df):
    weighted_math = df['UGDS'] * df['SATMTMID']
    return int(weighted_math.sum() / df['UGDS'].sum())
````

````python
college2.groupby('STABBR').apply(weighted_math_average).head()
````

If you try to do the above operation with the .agg() method, it overwrites value of every non-grouping column with SAT Math average --> .apply() shows its flexibility

.apply() also allows you to create multiple new columns by returning a Series. Let's beef up our function and get our desired results:

````python
from collections import OrderedDict
````

````python
def weighted_average(df):
    #Data container creation
    data = OrderedDict()
    #Raw total SATs
    weight_m = df['UGDS'] * df['SATMTMID']
    weight_v = df['UGDS'] * df['SATVRMID']
    #Calculate weighted mean
    wm_avg = weight_m.sum() / df['UGDS'].sum()
    wv_avg = weight_v.sum() / df['UGDS'].sum()
    #Assign new columns
    data['weighted_math_avg'] = wm_avg
    data['weighted_verbal_avg'] = wv_avg
    data['math_avg'] = df['SATMTMID'].mean()
    data['verbal_avg'] = df['SATVRMID'].mean()
    data['count'] = len(df)
    
    return pd.Series(data, dtype = 'float').round(0)
````

````python
college2.groupby("STABBR").apply(weighted_average).head(10)
````

## Find the longest streak of on-time flights

Find longest consecutive streak of on-time flights for each airline at eahc airport. Need to have each value in a column 'see' value ahead of it to find streaks

Will do this with `.diff` and `.cumsum` methods, before grouping by airline and airport

Begin with practice of counting:

````python
s = pd.Series([0, 1, 1, 0, 1, 1, 1, 0])
````

````python
s
````

````python
s1 = s.cumsum()
````

````python
s1
````

Now multiply streak dataset by original to get total cumulative:

````python
s.mul(s1)
````

GOing forward, we need to restart each streak at one. We can do this by chaining `.diff` method, 

then propogate negative values (end of streaks) down the series to subtract accumulation. 

Finally, we can propogate and merge the datasets

````python
s.mul(s1).diff()
````

````python
s.mul(s1).diff().where(lambda x: x < 0)
````

````python
s.mul(s1).diff().where(lambda x: x < 0).ffill()
````

````python
s.mul(s1).diff().where(lambda x: x < 0).ffill()\
 .add(s1, fill_value = 0)
````

Now we can find streaks! A bit complex but we can apply it easily to flights dataset, and then group

````python
flights = pd.read_csv('data/flights.csv')
````

````python
flights['ON_TIME'] = flights['ARR_DELAY'].lt(15).astype(int)
flights[['AIRLINE', 'ORG_AIR', 'ON_TIME']].head(10)
````

Define function we came up with in line 427:

````python
def max_streak(s):
    s1 = s.cumsum()
    return s.mul(s1).diff().where(lambda x: x < 0).ffill().add(s1, fill_value = 0).max()
````

Finally, calculate by grouping according to airport and airline, and sorting according to year and departure time to get consecutive counts

````python
flights.sort_values(['MONTH', 'DAY', 'SCHED_DEP']) \
       .groupby(['AIRLINE', 'ORG_AIR'])['ON_TIME'] \
       .agg(['mean', 'size', max_streak]).round(2).head()
````

## Stacking Multiple Groups of Variable Simultaneously

Some datasets contain multiple variable groups as columns anmes, and you need to melt them simultaneously. Let's see what we are talking about:

````python
movie = pd.read_csv('data/movie.csv')
actor = movie[['movie_title', 'actor_1_name',
               'actor_2_name', 'actor_3_name',
               'actor_1_facebook_likes', 
               'actor_2_facebook_likes', 
               'actor_3_facebook_likes']]
````

````python
actor.head()
````

Likely, we want to stack this dataset to 3 vairables:

1) Movie Title

2) Actor Name

3) Facebook Likes

However, we would ahve to stack the names and likes variables simultaneously, need a new method for this: `wide_to_long()`. To use this function, need to change column names to have them end in a digit:

````python
def change_col_names(col_name):
    col_name = col_name.replace('_name','')
    if 'facebook' in col_name:
        fb_idx = col_name.find('facebook')
        col_name = col_name[:5] + col_name[fb_idx - 1:] \
                                + col_name[5:fb_idx -1]
    return col_name
````

````python
actor2 = actor.rename(columns = change_col_names)

actor2.head()
````

Now that there is a linking key between the two column groups, use `.wide_to_long()` function to stack:

````python
stubs = ['actor', 'actor_facebook_likes']

actor_2_tidy = pd.wide_to_long(actor2, 
                               stubnames = stubs, 
                               i = ['movie_title'], 
                               j = 'actor_num', 
                               sep = '_')
````

````python
actor_2_tidy.head(10)
````

This method has a very specific input style. `stubnames` parameter accepts a list of strings, where each string represents a single column grouping. ALl columns that begin with this string will be stacked into a single column. However, in order to preserve the original identity of the multiple variables in one grouping, each column needs to end in a digit. We also need to tell `.wide_to_long()` that this digit is located after a `_` seperator. Finally, provide name for ending digit column `j`. 

We cfreated a function to define column names autmatically. First, the function `change_col_names()` removes `_name` from the end of each actor column. Next, it reorganizes the facebook variables to have the numeric digit placed at the end.
